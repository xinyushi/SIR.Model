So I tried to mimic the situation for elemetary schools, giving the average number of kids are 473, recover rate given by a poisson dist of lambda = 1/(13.61+13.24), and trying to varying b and see the infected people at peak day for example (this will be my plot 2). And the other plot is SIR model given below. 

```python
### Plot 1

## CONTINUOUS 

NumPop = 473 # Elemetary https://nces.ed.gov/programs/digest/d07/tables/dt07_095.asp
b = 1          # number of closed contact people 
k = 0.074    # 1/(13.61+13.24) for 0-19 people "https://www.ncbi.nlm.nih.gov/pmc/articles/PMC7229949/"
time=30      
ii = 0.02

sol1=SIR_continuous(b,k,time,ii)
plt.plot(sol1.t, sol1.y[0]*NumPop, c='b',label='s')    # generate the plot
plt.plot(sol1.t, sol1.y[1]*NumPop, c='g',label='r')
plt.plot(sol1.t, sol1.y[2]*NumPop, c='r',label='i')


plt.title("Continuous SIR Model(b=1, k=0.074 )")
plt.ylabel("Number of People")
plt.xlabel("Time Period")
plt.legend(['s','r','i'])
plt.savefig('C_SIR_1')

#print(sol1.y)
```
